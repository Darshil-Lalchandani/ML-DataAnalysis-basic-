```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Darshil\Desktop\python\appstore_games.csv")
```

```python
df.shape
```

```python
df.drop(["URL" , "ID" , "Subtitle" , "Description" , "Age Rating" , "Languages" , "Icon URL" , "Developer"],axis = 1,inplace = True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset = ["Average User Rating" ,"User Rating Count" ])
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df[df["User Rating Count"] > 200]
```

```python
df.shape
```

```python
from datetime import date as date
from datetime import datetime as dt
```

```python
df["Original Release Date"] = pd.to_datetime(df["Original Release Date"])
df["Current Version Release Date"] = pd.to_datetime(df["Current Version Release Date"])
```

```python
a = "01-10-2019"
a = pd.to_datetime(a)
```

```python
df["Upgrades"] = (df["Current Version Release Date"] - a).dt.days
```

```python
df["Upgrades"].value_counts()
```

```python
df = df[df["Upgrades"] < 0]
```

```python
df.shape
```

```python
df["Upgrades"].head(10)
```

```python
#Data cleaned completely(removed all null values,less user rating,recent update)
```

```python
df["Genres"].shape
```

```python
df.Genres=df.Genres.str.replace(', Entertainment','')
df.Genres=df.Genres.str.replace(', Games','')
df.Genres=df.Genres.str.replace('Games, ','')
df.Genres=df.Genres.str.replace('Entertainment, ','')
```

```python
df.Genres=df.Genres.str.replace('Strategy, ','')
df.Genres=df.Genres.str.replace(', Strategy','')
```

```python
df["Genres"].head(10)
```

```python
df.Genres = df.Genres.str.replace("Board","Puzzle",regex = True)
df.Genres = df.Genres.str.replace("Role Playing","Adventure",regex = True)
df.Genres = df.Genres.str.replace("Role","Adventure",regex = True)
df.Genres = df.Genres.str.replace("Education","Family",regex = True)
```

```python
df.Genres.value_counts(dropna = False)
```
